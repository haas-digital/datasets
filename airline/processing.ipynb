```python
ca_airports = ids[ids['Description'].str.contains("CA")]
ca_airports
```

```python
ca_df = df[(df['DEST_AIRPORT_ID'].isin(ca_airports['Code']))]
```

```python
carriers = pd.read_csv('L_UNIQUE_CARRIERS.csv')
carriers
```

```python
ca_carriers = carriers[carriers['Code'].isin(ca_df['MKT_UNIQUE_CARRIER'].unique())]
ca_carriers
```

```python
ca_carriers = ca_carriers.set_index(keys=ca_carriers['Code']).drop(columns=['Code'])
```

```python
ca_carriers_dict = ca_carriers.to_dict(orient='index')
```

```python
def convert_carrier_name(entry):
    return ca_carriers_dict[entry]['Description']
```

```python
ca_df['Carrier'] = ca_df['MKT_UNIQUE_CARRIER'].apply(convert_carrier_name)
```

```python
ca_df = ca_df.drop(columns=['Unnamed: 29', 'MKT_UNIQUE_CARRIER', 'OP_UNIQUE_CARRIER'])
```

```python
ca_df = ca_df.drop(columns=['ORIGIN_CITY_MARKET_ID', 'ORIGIN_AIRPORT_SEQ_ID', 'DEST_AIRPORT_SEQ_ID'])
```
