```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2021jan.csv')
```

```python
df = df.drop(columns=['Unnamed: 25'])
df.head()
```

```python
SFO_id = 14771
sfo_df = df[df['DEST_AIRPORT_ID'] == SFO_id]
```

```python
ids = pd.read_csv('L_AIRPORT_ID.csv')
```

```python
ca_airports = ids[ids['Description'].str.contains("CA")]
ca_airports
```

```python
# creating binary variable - 1 if the origin airport is in California, otherwise 0
from_ca = np.where(sfo_df['ORIGIN_AIRPORT_ID'].isin(ca_airports['Code']), 1, 0)
```

```python
sfo_df['origin_CA'] = from_ca
```

## OMIT

The following cells decoded the carrier code for each flight - this won't be used in the final dataset.

```python
carriers = pd.read_csv('L_UNIQUE_CARRIERS.csv')
ca_carriers = carriers[carriers['Code'].isin(sfo_df['MKT_UNIQUE_CARRIER'].unique())]
ca_carriers
```

```python
ca_carriers = ca_carriers.set_index(keys=ca_carriers['Code']).drop(columns=['Code'])
```

```python
ca_carriers_dict = ca_carriers.to_dict(orient='index')
```

```python
def convert_carrier_name(entry):
    return ca_carriers_dict[entry]['Description']
```

```python
ca_df['Carrier'] = ca_df['MKT_UNIQUE_CARRIER'].apply(convert_carrier_name)
```
